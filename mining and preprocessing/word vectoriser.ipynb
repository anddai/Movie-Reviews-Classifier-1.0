```python
import pandas as pd
import os
import ast
```

```python
path = os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir) + "/data")

print(os.listdir(path))

reviews = pd.read_csv(path + '/raw/movie_reviews.csv')

word_vectors_df = pd.DataFrame(columns=["POSITIVE","NEGATIVE"])
```

```python
def vectorise_rating(review_text, sentiment, word_vectors_df):
    for word in review_text: 
        if word not in list(word_vectors_df.index):
            word_vectors_df.loc[word] = [0,0]
        word_vectors_df.at[word, sentiment] += 1
    return word_vectors_df

for review in reviews.itertuples(name='Pandas'):
    review_text = ast.literal_eval(review.review_text)
    sentiment = review.sentiment
    word_vectors_df = vectorise_rating(review_text, sentiment, word_vectors_df)

#Score is (Positive - Negative) / (Positive + Negative + 1)?
```
