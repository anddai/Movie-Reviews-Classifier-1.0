```python
# import essential libraries 
import pandas as pd
import os
```

```python
# read in data
path = os.path.join(os.path.normpath(os.getcwd() + os.sep + os.pardir) + "/data")

print(os.listdir(path))

reviews = pd.read_csv(path + '/cleaned/word_vectors.csv')
```

```python
tot_pos = reviews["POSITIVE"].sum()
tot_neg = reviews["NEGATIVE"].sum()

print(tot_pos/tot_neg)

# adjust for unbalance in data and weight based on strength of data 
reviews["POSITIVE"] += 1
reviews["NEGATIVE"] = reviews["NEGATIVE"] * tot_pos/tot_neg + 1
```

```python
# Calculate Bayes Probability Scores, adjusted for unbalance in data, deweighted based on strength
reviews["POS_BAYES"] = reviews["POSITIVE"]/(reviews["POSITIVE"] + reviews["NEGATIVE"])
reviews["NEG_BAYES"] = reviews["NEGATIVE"]/(reviews["POSITIVE"] + reviews["NEGATIVE"])
```

```python
reviews = reviews.set_index('Unnamed: 0')
reviews = reviews[["POS_BAYES", 'NEG_BAYES']]

reviews

#reviews.to_csv(path + '/cleaned/naivebayes.csv', header=True, index=True)
```
